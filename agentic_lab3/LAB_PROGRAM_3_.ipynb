# Human-in-the-Loop: Approve Before Sending

```python
from google.colab import userdata
from langchain_groq import ChatGroq

groq_api_key = userdata.get("GROQ_API_KEY")
llm = ChatGroq(api_key= groq_api_key,
               model="llama-3.1-8b-instant", temperature=0.3)
```

```python
task = 'Write a short polite email asking my professor for a deadline extension.'
draft = llm.invoke(task).content
print('----- DRAFT EMAIL -----')
print(draft)
```

```python
decision = input('Type APPROVED to send, or anything else to reject: ')
if decision.strip().upper() == 'APPROVED':
    print('✅ Email sent!')   # in a real app, sending happens here
else:
    print('❌ Not sent. The draft was rejected.')
```
